```python
import copy
import json
import pandas as pd
import spacy
from sklearn.feature_extraction.text import CountVectorizer
```

```python
filename = "./JobsCountVecVocab.txt"

with open(filename) as f:
    data = f.read()
f.close()
jobsVocab = json.loads(data)

filename = "./ResumeCountVecVocab.txt"
with open(filename) as f:
    data = f.read()
f.close()
resumeVocab = json.loads(data)
```

```python
print(jobsVocab)
```

```python
print(resumeVocab)
```

```python
overlap = {}
count = 0
for jobs in jobsVocab:
    for i, vocab in enumerate(resumeVocab):
        if jobs == vocab:
            overlap[jobs] = min(jobsVocab.get(jobs),resumeVocab.get(vocab))
            # count += 1
            # print("Jobs: " + jobs + ": " + str(jobsVocab.get(jobs)))
            # print("Vocab " + str(i) + ": " + vocab + ": " + str(resumeVocab.get(vocab)))
            
            
```

```python
print(count)
print(len(jobsVocab))
print(len(resumeVocab))
print((overlap))

```

```python
vectorizer = CountVectorizer(min_df=1)
vectorizer.fit(overlap)
vector = vectorizer.transform(overlap)
vector_features = vectorizer.get_feature_names_out()
vector_arr = vector.toarray()
count_vect_df = pd.DataFrame(vector.todense(), columns=vector_features)
```

```python
print("Vocabulary: ", len(vectorizer.vocabulary_))
print(f"Found {len(vectorizer.get_feature_names_out())} unique terms")
print("Encoded Document is:")
print(vector)
vectorizer.vocabulary_
```
